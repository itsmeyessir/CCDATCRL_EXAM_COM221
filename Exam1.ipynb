## Search

```python
def linearSearch(my_list, number):

    index = 0

    for i in my_list:

        if my_list[index] == number:
            return index

        index = index + 1

    return -1
```

1. Using **linear search**, write a `findLargestIncrease()` function to find the largest increase in value between two adjacent numbers in your student number `(10 points)`

    **Example:**

    student_number = [2,0,2,2,6,2,5,1,9,3]

    **Expected Output:**

    findLargestIncrease(student_number)

    Output: Between 1 and 9

    **Explanation:**

    From this student_number, the largest increase happened from number 1 to number 9

```python
# put your answer here
def findLargestIncrease(student_number):
    if len(student_number) < 2:
        return "No adjacent elements to compare"

    largest_increase = 0
    start_index = 0
    end_index = 0

    for i in range(len(student_number) - 1):
        current_increase = student_number[i + 1] - student_number[i]

        if current_increase > largest_increase:
            largest_increase = current_increase
            start_index = linearSearch(student_number, student_number[i])
            end_index = linearSearch(student_number, student_number[i + 1])

    if largest_increase > 0:
        return f"Between {student_number[start_index]} and {student_number[end_index]}"
    else:
        return "No increase found in adjacent numbers"

student_number = [2, 0, 2, 2, 1, 0, 3, 3, 1, 0]
result = findLargestIncrease(student_number)
print("Student Number:", student_number)
print("Output:", result)
print("Explanation: From this student number, the largest increase happened from between 0 to 3")
```

## Queue

```python
class Queue:
    def __init__(self):
        self.queue = []

    # Add an element
    def enqueue(self, item):
        self.queue.append(item)

    # Remove an element
    def dequeue(self):
        if len(self.queue) < 1:
            return None
        return self.queue.pop(0)

    # Display  the queue
    def display(self):
        print(self.queue)

    def size(self):
        return len(self.queue)

    def peek(self):
        return self.queue[0]
```

4. Write a `removeDup()` function that will remove all duplicates in **num_queue** `(5 points)`

    **Expected Output:**

    removeDup(num_queue)

    [2,0,1,3,4,5,6]

```python
# put your answer here
def removeDup(queue):
    unique_items = set()
    new_queue = Queue()

    for item in queue.queue:
        if item not in unique_items:
            unique_items.add(item)
            new_queue.enqueue(item)

    queue.queue = new_queue.queue

student_number = "2022103310"

num_queue = Queue()

for digit in student_number:
    num_queue.enqueue(int(digit))

removeDup(num_queue)

print("removeDup(num_queue)")
num_queue.display()
```

1. Using any sorting algorithm, sort the **last 6 digits** of your student number in ascending order `(5 points)`

    **Example:**

    student_number = [2,0,2,2,6,2,5,1,3,4]

    **Expected Output:**

    print(student_number)

    [2,0,2,2,1,2,3,4,5,6]


```python
# put your answer here
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    left = merge_sort(left)
    right = merge_sort(right)

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

student_number = [2, 0, 2, 2, 1, 0, 3, 3, 1, 0]

sorted_last_six_digits = merge_sort(student_number[-6:])
student_number[-6:] = sorted_last_six_digits

print("print(student_number)")
print(student_number)
```

2. Using your student_number, separately sort the even numbers in **increasing order** and the odd numbers in **decreasing order** `(15 points)`

    **Example:**

    student_number = [2,0,2,2,1,2,3,4,5,6]

    **Expected Output:**

    print(student_number)

    [0,2,2,2,2,4,6,5,3,1]

```python
# put your answer here
def merge_sort(arr, reverse=False):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    left = merge_sort(left, reverse)
    right = merge_sort(right, reverse)

    return merge(left, right, reverse)

def merge(left, right, reverse=False):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if (left[i] < right[j]) ^ reverse:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

student_number = [2, 0, 2, 2, 1, 0, 3, 3, 1, 0]

even_numbers = [x for x in student_number if x % 2 == 0]
odd_numbers = [x for x in student_number if x % 2 != 0]

even_numbers_sorted = merge_sort(even_numbers)
odd_numbers_sorted = merge_sort(odd_numbers, reverse=True)

student_number = even_numbers_sorted + odd_numbers_sorted

print(student_number)
```
